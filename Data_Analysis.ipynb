```python
import pandas as pd

df = pd.read_csv('/content/sample_data/california_housing_train.csv')
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df['median_house_value'], bins=20, edgecolor='black')
plt.title('Distribution of Median House Value')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(df['longitude'], df['latitude'], alpha=0.5)
plt.title('Latitude vs Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python
ys = df['median_house_value'].to_numpy()
xs = df.drop('median_house_value', axis=1).to_numpy()
ys.shape, xs.shape
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(xs, ys, test_size=0.2, random_state=42)

print(f"Shape of x_train: {x_train.shape}")
print(f"Shape of x_test: {x_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

```python
# train model linear regression
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
#promt: evaluate model regression

import numpy as np
from sklearn.metrics import mean_squared_error

#evaluate model
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print('NSE =',mse)
print('RMSE =',rmse)
```

```python
index = 100

ys_pred = model.predict([x_test[index]])
print('ys_pred = ', ys_pred[0])
print('ys_test = ', y_test[index])
```
